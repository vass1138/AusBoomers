# Australian Men's Basketball Tokyo 2020

Comparison of historical efficiency with latest value in Tokyo 2020 Olympics.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def compute_epg(da):
    
    # missing TO
    # per-game & percentages available only
    da['EFF.Calc'] = da['Pts'] + da['Reb'] + da['Ast'] + da['Stl'] + da['Blk'] + da['FG%']/100.0 + da['3%']/100.0 + da['1%']/100

    return(da)
```

```python
df_players = pd.read_excel('../Data/aus_boomers_players.xlsx')

# Filter players to final team selection
df_players = df_players.loc[df_players['Selected'] == 'Y']

# Select first 7 rows
df_players = df_players.iloc[:, :7]

df_players.head()
```

```python
# df_stats = pd.read_excel('/content/drive/My Drive/AusBoomers/aus_boomers_proballer_stats.xlsx')
# df_stats.head()
```

```python
df_stats = pd.read_excel('../Data/aus_boomers_proballer_stats.xlsx')
df_stats.head()
```

```python
df_stats.info()
```

```python
# how does published Eff compare with ad-hoc EFF.Calc value?
da = compute_epg(df_stats)
```

```python
# construct year from season label
da['Year'] = "20" + da['Season'].str[:2]
da['Year'] = pd.to_numeric(da['Year'])
```

```python
da.head()
```

```python
dh = da[da['Source']=="list-regular"].Name.unique()
len(dh)
```

```python
# current regular season stats
df1 = da[da['Source']=="list-regular"]
df2 = df1.groupby(["Name"]).apply(lambda x: x.sort_values(["Year"], ascending = False)).reset_index(drop=True)
dc = df2.groupby('Name').head(1)
```

```python
dt = dc[['Season','Age','League','Team','Name']]
```

```python
# concatenate current team with player
dh = da[['Year','Season','Name','Eff']]
dh = dh.merge(dt,on='Name')

dp = df_players[['Name','Position','Height']]
dh = dh.merge(dp,on='Name')

dh['Player-Team'] = dh['Name'] + '\n(' + dh['Team'] + ') ' + '\n Pos: ' + dh['Position'] + ' '+ ' Height: ' + dh['Height'].astype('str') + ' '
```

```python
# calculate mean for given year
dh = dh.groupby(['Name','Year','Player-Team','Position'],as_index=False).mean()
```

```python
dh.Position.unique()
```

```python
# order player positions
t = pd.CategoricalDtype(categories=['PG', 'SG', 'SF', 'PF', 'C'], ordered=True)
dh['sort']=pd.Series(dh.Position, dtype=t)
dh.sort_values(by=['sort','Name','Height'],ascending=[True,True,True],ignore_index=True,inplace=True)
```

```python
dh.head()
```

## Olympic stats

```python
dt2020_stats = pd.read_excel('../Data/tokyo2020_proballer_stats.xlsx')
```

```python
dt2020_stats = compute_epg(dt2020_stats)

dt2020_stats['Diff'] = dt2020_stats['EFF.Calc'] - dt2020_stats['Eff']
dt2020_stats['Ratio'] = dt2020_stats['EFF.Calc'] / dt2020_stats['Eff']
dt2020_stats
```

```python
# sort facets by position and height
# sort data by year
dh = dh.sort_values(by=['sort','Height','Year'])
```

```python
# get the list of ordered names from the dataframe (collapsing duplicates)
ordered_names = pd.Series(dh['Name']).unique()

# reformat as dataframe (so we can do a join later)
dy = pd.DataFrame(ordered_names,columns=['Player'])
```

```python
# subset of Olympic stats
dz = dt2020_stats[['Player','EFF.Calc']]
dz.head()
```

```python
# sort efficiency values using our sorted names
dy = dy.join(dz.set_index('Player'),on='Player')[['Player','EFF.Calc']]
dy.head()
```

```python


# Create a grid : initialize it
g = sns.FacetGrid(dh, col='Player-Team', hue='Position', col_wrap=4)
 
# Add the line over the area with the plot function
g = g.map(plt.plot, 'Year', 'Eff')
 
# Fill the area with fill_between
g = g.map(plt.fill_between, 'Year', 'Eff', alpha=0.2).set_titles("{col_name} Player-Team")
 
# Control the title of each facet
g = g.set_titles("{col_name}")

axes = g.axes.flatten()

for i, ax in enumerate(axes):
    ax.axhline(dy.iloc[i]['EFF.Calc'], ls='--', c='purple')
 
# Add a title for the whole plot
plt.subplots_adjust(top=0.85)
sns.set(rc={'figure.figsize':(12,8)})
g = g.fig.suptitle('Australian Men\'s Olympic Basketball Performance\nHistorical Effciency vs Final Tokyo 2020 Olympic Value')

plt.text(2020,-18,"linkedin.com/in/evassiliadis",ha='right',alpha=0.5)
# plt.text(2020,-16,"github.com/vass1138/wnbl",ha='right',alpha=0.5)
plt.savefig('boomers_tokyo2020_eff_history.png')
plt.show()
```
